```python
from collections import defaultdict
eigen_centrality = defaultdict(int)
total_eigen_centrality = defaultdict(int)
mean_eigen_centrality = defaultdict(int)
betweeness_centrality = defaultdict(int)
total_betweeness_centrality = defaultdict(int)
mean_betweeness_centrality = defaultdict(int)
for i in range(1,12): 
    g = G[i]
    eigen_centrality[i] = nx.algorithms.centrality.eigenvector_centrality(g)
    betweeness_centrality[i] = nx.algorithms.centrality.betweenness_centrality(g)
    for k, v in eigen_centrality[i].items():
        total_eigen_centrality[k] += v

    for k, v in betweeness_centrality[i].items():
        total_betweeness_centrality[k] += v
        
for k, v in total_eigen_centrality.items():
    mean_eigen_centrality[k] = v / 11
    
for k, v in total_betweeness_centrality.items():
    mean_betweeness_centrality[k] = v / 11
```

```python
#file_name = "release_networks_homework_data/CAVIAR/" + var_name + ".csv"
phases2 = {}
G2 = {}
for i in range(1,12):
    var_name2 = "phase" + str(i)
    file_name2 = "https://raw.githubusercontent.com/ragini30/Networks-Homework/main/" + var_name2 + ".csv"
    phases2[i] = pd.read_csv(file_name2, index_col = ["players"])
    phases2[i].columns = "n" + phases2[i].columns
    phases2[i].index = phases2[i].columns
    G2[i] = nx.from_pandas_adjacency(phases2[i], create_using=nx.DiGraph())
    G2[i].name = var_name2
Res = {}
for num, g in G2.items():
    Res[num] = nx.algorithms.link_analysis.hits_alg.hits(g, max_iter=1000000)
```

```python
import matplotlib.pylab as plt
n1_hub = []
n1_authority = []
n3_hub = []
n3_authority = []

for k, v in Res.items():
    n1_hub.append(v[0]['n1']) 
    n1_authority.append(v[1]['n1'])
    n3_hub.append(v[0]['n3']) 
    n3_authority.append(v[1]['n3'])
    
list_of_tuples = list(zip(range(1,12), n1_hub, n1_authority, n3_hub, n3_authority))
df = pd.DataFrame(list_of_tuples, columns = ['x_values', 'n1_hub', 'n1_authority', 'n3_hub', 'n3_authority'])
plt.plot( 
    'x_values', 
    'n1_hub', 
    data=df, 
    marker='o', 
    markerfacecolor='blue', 
    markersize=6, 
    color='skyblue', 
    linewidth=3, 
    label="n1_hub"
)
plt.plot( 
    'x_values', 
    'n1_authority', 
    data=df, 
    marker='o', 
    markerfacecolor='green', 
    markersize=6, 
    color='olive', 
    linewidth=3, 
    label="n1_authority"
)
plt.plot(
    'x_values', 
    'n3_hub', 
    data=df, 
    marker='o', 
    markerfacecolor='orange', 
    markersize=6, 
    color='yellow', 
    linewidth=3, 
    label="n3_hub")
plt.plot( 'x_values', 'n3_authority', data=df, marker='o', markerfacecolor='red', markersize=6, color='pink', linewidth=3, label="n3_authority")
plt.legend()
plt.rcParams["figure.figsize"] = (25,10)
plt.legend(fontsize=16)
plt.xlabel('Phases', fontsize=18)
plt.ylabel('Hub or Auth Score', fontsize=18)
plt.show()
```

```python
import matplotlib.pylab as plt
n1_eigenvector_centrality = []
n1_betweenness_centrality = []
n1_degree_centrality = []
n3_eigenvector_centrality = []
n3_betweenness_centrality = []
n3_degree_centrality = []

for k, v in G.items():
    g = G[k]
    n1_eigenvector_centrality.append(nx.algorithms.centrality.eigenvector_centrality(g)['n1']) 
    n1_betweenness_centrality.append(nx.algorithms.centrality.betweenness_centrality(g)['n1']) 
    n1_degree_centrality.append(nx.algorithms.centrality.degree_centrality(g)['n1']) 
    n3_eigenvector_centrality.append(nx.algorithms.centrality.eigenvector_centrality(g)['n3']) 
    n3_betweenness_centrality.append(nx.algorithms.centrality.betweenness_centrality(g)['n3']) 
    n3_degree_centrality.append(nx.algorithms.centrality.degree_centrality(g)['n3']) 
    
list_of_tuples = list(zip(
    range(1,12), 
    n1_eigenvector_centrality, 
    n1_betweenness_centrality, 
    n1_degree_centrality, 
    n3_eigenvector_centrality,
    n3_betweenness_centrality,
    n3_degree_centrality
))
df = pd.DataFrame(
    list_of_tuples, 
    columns = [
        'x_values', 
        'n1_eigenvector_centrality', 
        'n1_betweenness_centrality', 
        'n1_degree_centrality', 
        'n3_eigenvector_centrality', 
        'n3_betweenness_centrality',
        'n3_degree_centrality'
    ]
)

plt.plot( 
    'x_values', 
    'n1_eigenvector_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='blue', 
    markersize=6, 
    color='skyblue', 
    linewidth=3, 
    label="n1_eigenvector_centrality"
)
plt.plot( 
    'x_values', 
    'n1_betweenness_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='green', 
    markersize=6, 
    color='olive', 
    linewidth=3, 
    label="n1_betweenness_centrality"
)
plt.plot( 
    'x_values', 
    'n1_degree_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='purple', 
    markersize=6, 
    color='plum', 
    linewidth=3, 
    label="n1_degree_centrality"
)
plt.plot(
    'x_values', 
    'n3_eigenvector_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='orange', 
    markersize=6, 
    color='yellow', 
    linewidth=3, 
    label="n3_eigenvector_centrality")
plt.plot(
    'x_values', 
    'n3_betweenness_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='red', 
    markersize=6, 
    color='pink', 
    linewidth=3, 
    label="n3_betweenness_centrality"
)
plt.plot(
    'x_values', 
    'n3_degree_centrality', 
    data=df, 
    marker='o', 
    markerfacecolor='burlywood', 
    markersize=6, 
    color='peachpuff', 
    linewidth=3, 
    label="n3_degree_centrality"
)

plt.legend()
plt.rcParams["figure.figsize"] = (25,10)
plt.legend(fontsize=16)
plt.xlabel('Phases', fontsize=18)
plt.ylabel('Centrality', fontsize=18)
plt.show()
```
